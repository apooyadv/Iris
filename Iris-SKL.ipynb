```python
import seaborn as sns
import pandas as pd
```

```python
pd.set_option('display.max_rows',150)
iris_df = sns.load_dataset('iris')
sns.set_style('darkgrid')
```

```python
iris_df
```

```python
Len_Plot = sns.scatterplot(iris_df,x='sepal_length',y='petal_length',hue='species').set(title='Petal Len Vs Sepal Len')
```

```python
sns.scatterplot(iris_df,x='sepal_width',y='petal_width',hue='species').set(title='Petal Width Vs Sepal Width')
```

```python
iris_df.columns
```

```python
num_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = iris_df[num_cols]
```

```python
X
```

**KMEANS**

```python
from sklearn.cluster import KMeans 
```

```python
model = KMeans(n_clusters=3,random_state=42)
model.fit(X)
```

```python
sns.scatterplot(iris_df,x='sepal_length',y='petal_length',hue='species').set(title='Petal Len Vs Sepal Len')
```

```python
preds = model.predict(X)
preds
```

```python
sns.scatterplot(iris_df,x='sepal_length',y='petal_length',hue=preds).set(title='Petal Len Vs Sepal Len')
```

```python
X.columns
```

```python
model.cluster_centers_
```

```python
model.inertia_
```

```python
import matplotlib.pyplot as plt
```

```python
range = range(2,11)
inertia = []

for n_clusters in range:
    model = KMeans(n_clusters=n_clusters,random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)
```

**DBSCAN**

```python
plt.plot(range,inertia,'-o')
plt.title('Inertia Vs N_Clusters')
plt.xlabel('N_Cluster')
plt.ylabel('Inertia')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.5,min_samples=3)
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
sns.scatterplot(X,y='petal_length',x='sepal_length',hue=model.labels_).set(title='DBSCAN Preds')
```

**SOME METHODS USING DIMENSIONALITY REDUCTION**

PRINCIPAL COMPONENT ANALYSIS

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
transformed = pca.transform(X)
```

```python
transformed[:, 1]
```

```python
sns.scatterplot(y=transformed[:, 0],x=transformed[:, 1],hue=iris_df['species']).set(title='PCA')
```

t-SNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2)
```

```python
transformed = tsne.fit_transform(X)
```

```python
sns.scatterplot(x=transformed[:,0],y=transformed[:,1],hue=iris_df['species']).set(title='TSNE')
```
